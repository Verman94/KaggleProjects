```python
def summary(df):
    print(f"Dataset has {df.shape[1]-1} features and {df.shape[0]} examples.")
    summary = pd.DataFrame(index=df.columns)
    desc = pd.DataFrame(df.describe(include='all').transpose())
    summary["Count"] = desc['count'].values
    summary["Unique"] = df.nunique().values
    summary["Missing"] = df.isnull().sum().values
    summary["Duplicated"] = df.duplicated().sum()
    summary['Std'] = desc['std'].values
    summary["Mode"] = df.mode().values[0]
    summary["Median"] = df.median()
    summary['Mean'] = desc['mean'].values
    summary['Min'] = desc['min'].values
    summary['Max'] = desc['max'].values
    summary["First Value"] = df.loc[0].values
    summary["Last Value"] = df.loc[df.shape[0]-1].values
    summary["Types"] = df.dtypes
    return summary
```

```python
# Removing outliers
def outlier_removal(data , i):
    q1 = np.quantile(data[i] ,0.25)
    q3 = np.quantile(data[i] , 0.75)
    iqr = q3-q1
    lower_tail = q1 - 1.5*iqr
    upper_tail = q3 + 1.5*iqr
    data.drop(data[data[i]> upper_tail].index , inplace = True)
    data.drop(data[data[i]< lower_tail].index ,inplace =True)
    
outlier_list = ['honeybee', 'bumbles', 'osmia', 'fruitset', 'fruitmass', 'RainingDays', 'seeds']
nonoutliers_df=comb_df.copy()
for  i in outlier_list:
    outlier_removal(nonoutliers_df ,i)
```

```python
def getXy(train, test, y_label, X_label=None,ScaleLabels=None):
#     scaler = MinMaxScaler()
    scaler = StandardScaler()
    X_sub,X,y = [],[],[]
    if X_label is None:
        features = list(test.columns)
        X = train[features]
        X_sub = test[features]
        if ScaleLabels is not None:
            X[ScaleLabels] = scaler.fit_transform(X[ScaleLabels])
            X_sub[ScaleLabels] = scaler.fit_transform(X_sub[ScaleLabels])

    else:
        if len(X_label) == 1:
            features = X_label
            X = train[X_label].values.reshape(-1,1)
            X_sub = test[X_label].values.reshape(-1,1)
            if ScaleLabels is not None:
                X[ScaleLabels] = scaler.fit_transform(X[ScaleLabels])
                X_sub[ScaleLabels] = scaler.fit_transform(X_sub[ScaleLabels])
        else:
            features = X_label
            X = train[X_label].copy()
            X_sub = test[X_label].copy()
            if ScaleLabels is not None:
                X[ScaleLabels] = scaler.fit_transform(X[ScaleLabels])
                X_sub[ScaleLabels] = scaler.fit_transform(X_sub[ScaleLabels])
    y = train[y_label]
#     y = train[y_label].values.reshape(-1,1)
    return X_sub,X,y,features
```

In this technique, the data in the testing set is compared against the training set, and the algorithm attempts to identify examples in the testing set that are similar to the training set. This technique can be useful in identifying potential problems with data before training a model and can help prevent overfitting.
Adversarial Validation is a very clever and very simple way to let us know if our test data and our training data are similar; we combine our train and test data, labeling them with say a 0 for the training data and a 1 for the test data, mix them up, then see if we are able to correctly re-identify them using a binary classifier.

```python
# TRAIN
def runXgboost(train, test, y_label, n_splits, params, X_label=None):
    
    # Get X and y and X_sub
    X_sub,X,y,features = getXy(train, test, y_label, X_label=X_label)
    # Define cross validation model
    kfold = KFold(shuffle = True, n_splits=n_splits, random_state=Seed)
    # Store feature importance
    labels = pd.Index(features)
    importance_xgb = []
    # Store model scores and oof preds
    model_scores,rmse_scores, acc_scores = [],[],[]
    val_preds = np.zeros(len(X))
    test_preds = np.zeros(test.shape[0])

    for fold_idx, (train_idx, val_idx) in enumerate(kfold.split(X,y)):
        X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
        y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]
        # requires xgboost.DMatrix.
        dtrain = xgb.DMatrix(X_train, label=y_train, feature_names=features)
        dval = xgb.DMatrix(X_val, label=y_val, feature_names=features)
        watchlist = [(dtrain, 'train'), (dval, 'eval')]
        early_stop = xgb.callback.EarlyStopping(rounds=params['early_stopping_rounds'], metric_name='mae', data_name='eval', save_best=True)
        model = xgb.train(params, dtrain, num_boost_round=params['n_estimators'], evals=watchlist, verbose_eval=False, callbacks=[early_stop])

        importance_xgb.append(model.get_score(importance_type='gain'))
        
        val_preds[val_idx] = model.predict(dval)
        test_preds += model.predict(xgb.DMatrix(X_sub, feature_names=features)) / (n_splits)
        
        # Evaluation different metrics
        mae = model.eval(dval).split(':')[1].strip()    
        mae = float(mae)
        accuracy = r2_score(y_val, val_preds[val_idx])
        rmse = mean_squared_error(y_val, val_preds[val_idx], squared=False)
        acc_scores.append(accuracy)
        model_scores.append(mae)
        rmse_scores.append(rmse)
#         print(dval[3])
        # Print metrics
        print(f'Fold {fold_idx+1}: MAE score = {mae:.5f}     RMSE scores = {rmse:.5f}       R2 scores = {accuracy:.5f}')
    
    mean_scores = np.mean(model_scores)
    mean_acc_scores = np.mean(acc_scores)
    mean_rmse_scores = np.mean(rmse_scores)
    print(f'{"_" * 80}\n Mean MAE score: {mean_scores:.4f} (+/- {np.std(model_scores)**2})  Mean RMSE score: {mean_rmse_scores:.4f}    Mean R2 score: {mean_acc_scores:.4f}')
    
    return features,mean_scores,mean_rmse_scores,test_preds, val_preds, importance_xgb
```

```python
# SKlearn wrapper
def runXgboostfit(train, test, y_label, n_splits, params, X_label=None):
    
    # Get X and y and X_sub
    X_sub,X,y,features = getXy(train, test, y_label, X_label=X_label)
    # Define cross validation model
    kfold = KFold(shuffle = True, n_splits=n_splits, random_state=Seed)
    # Store feature and permutation importance
    labels = pd.Index(features)
    importance_xgb,importance_train,importance_val = [],[],[]
    # Store model scores and oof preds
    model_scores,acc_scores = [],[]
    test_preds = np.zeros((test.shape[0]))

    for fold_idx, (train_idx, val_idx) in enumerate(kfold.split(X,y)):
        X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
        y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]
        model = XGBRegressor(**params)
        early_stop = xgb.callback.EarlyStopping(rounds=params['early_stopping_rounds'], metric_name='mae', save_best=True)
        model.fit(X_train, y_train, eval_set=[(X_val, y_val)], verbose=False, callbacks=[early_stop])
        importance_xgb.append(model.feature_importances_)   
        y_pred = model.predict(X_val)
        test_pred = model.predict(X_sub)
        # return   
        test_preds += test_pred / (n_splits)
        
        # Evaluation different metrics
        accuracy = model.score(X_val, y_val)
        mae = mean_absolute_error(y_val,y_pred)
        model_scores.append(mae)
        acc_scores.append(accuracy)
        
        # Print metrics
        print(f'Fold {fold_idx+1}: MAE score = {mae:.5f}     R2 scores = {accuracy:.5f}')
    
    mean_scores = np.mean(model_scores)
    mean_acc_scores = np.mean(acc_scores)
    print(f'{"_" * 80}\n Mean MAE score: {mean_scores:.4f} (+/- {np.std(model_scores)**2})  Mean R2 score: {mean_acc_scores:.4f}')
    return features,mean_scores,mean_acc_scores,test_preds,importance_xgb
```

```python
def MAE_score_model_fit(model, X_sub, X, y, cv, totalFolds, verbose=True):
    test_preds = np.zeros(len(X_sub))
    scores=[]
    for fold_idx, (train_idx, val_idx) in enumerate(cv.split(X,y)):
        X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
        y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]
        model.fit(X_train, y_train)
        y_pred = model.predict(X_val)
        test_pred = model.predict(X_sub)
        test_preds += test_pred / (totalFolds)
        
#         # test Mat OP suggestion
#         y_pred = mattop_post_process(y_pred)
#         test_preds = mattop_post_process(test_preds)
       
        # Metrics
        score = mean_absolute_error(y_val, y_pred)
        scores.append(score)
        accuracy = model.score(X_val, y_val)
        if verbose:
            print(f'Fold {fold_idx+1}: MAE score = {score:.5f}     R2 scores = {accuracy:.5f}')
    return test_preds,scores
```

```python
def multimodelEval(Models, ):
    '''
    This function is to evaluate diffrent models that are fed as an dictionary to find the best model
    based on the requiered metrics sorting is in Ascending order and double check the print of the best model
    return of the function are 
    "sorted_names":    Sorted model names
    "sorted_mean_scores":   Mean score of models from each fold sorted
    "finalPred": Dataframe which is created from "predictions" An array to store values of test_preds to submit
    '''
    # store models score and test set preds
    predictions = []
    scores = []
    for name, model in Regressors.items():
        model_scores = []
        test_preds,score = MAE_score_model_fit(model, X_sub, X, y, kfold, totalFolds=n_splits)

        
        model_scores.append(score)

        # Add score of the model to the list
        scores.append((name, model_scores))
        predictions.append((name, test_preds))
        
        # Print classification score of the model
        print(f'\n Model is "{name}"  MAE score is {np.mean(model_scores):.5f} \n{"*" * 80}')
    
    # Find mean score of models and sort them and print the best model
    names, model_scores = zip(*scores)
    mean_scores = [np.mean(s) for s in model_scores]
    comp = dict(zip(names,mean_scores))
    sorted_comp = dict(sorted(comp.items(), key=lambda x: x[1]))
    sorted_names = list(sorted_comp.keys())
    sorted_mean_scores = list(sorted_comp.values())
    print(f'{"_" * 80}\n The best model is "{sorted_names[0]}" with a mean MAE score of {sorted_mean_scores[0]:.4f}')
    
    # store predictions in dataframe
    modelNames, predicts = zip(*predictions)
    combine = dict(zip(modelNames,predicts))
    finalPred = pd.DataFrame(combine)
    
    return sorted_names,sorted_mean_scores,finalPred
```
