```python
def summary(df):
    print(f'{Style.BRIGHT}{Fore.BLACK}Head of the dataset \n \n {"*"*100}')
    display(df.head(3))
    print(f'\n {Style.BRIGHT}{Fore.BLACK}{"*"*100}\n')
    print(f"{Style.BRIGHT}{Fore.BLACK}Summary of the dataset ---->  dataset has {Fore.RED}{df.shape[1]-1}{Fore.BLACK} features and {Fore.RED}{df.shape[0]}{Fore.BLACK} examples.")
    print(f'\n{Style.BRIGHT}{Fore.BLACK}{"*"*100}\n{Style.RESET_ALL}')
    summary = pd.DataFrame(index=df.columns)
    desc = pd.DataFrame(df.describe(include='all').transpose())
    summary["Count"] = desc['count'].values
    summary["Unique"] = df.nunique().values
    summary["Missing"] = df.isnull().sum().values
    summary["Duplicated"] = df.duplicated().sum()
    summary['Std'] = desc['std'].values
    summary["Mode"] = df.mode().values[0]
    summary["Median"] = df.median()
    summary['Mean'] = desc['mean'].values
    summary['Min'] = desc['min'].values
    summary['Max'] = desc['max'].values
    summary["First Value"] = df.loc[0].values
    summary["Last Value"] = df.loc[df.shape[0]-1].values
    summary["Types"] = df.dtypes
    return display(summary)
```

```python
def piePlot(ax, data, plot_params, adjust_labels=None):
    
    '''
    adjust_labels : Inputs a list of constants to put each label in an desiered y-axis position
    
    plot a pie plot with custom annotation:
    
    text : The text of the annotation
    xy : The point (x,y) to annotate
    xytext : The position (x,y) to place the text at (If None, defaults to xy)
    arrowprops : The properties used to draw an arrow between the positions xy and xytext
    
    '''
    
    labels = list(data.index)
    wedges, texts, autotexts = ax.pie(data, **plot_params)
    
    # store pct values
    autopct_values = [autotext.get_text() for autotext in autotexts]
    
    # Remove autopct values from the chart
    for autotext in autotexts:
        autotext.set_visible(False)
        
    # Create boxes and arrow for labels
    bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
    kw = dict(arrowprops=dict(arrowstyle="->", connectionstyle='arc3,rad=0.5'),
    bbox=bbox_props, zorder=0, ha='center', va='center', fontsize=8, weight='bold')
    
    # Using Annotate for labeling
    for i, (p, label) in enumerate(zip(wedges, labels)):
        ang = (p.theta2 - p.theta1)/2. + p.theta1
        y = np.sin(np.deg2rad(ang))
        x = np.cos(np.deg2rad(ang))
        horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
        connectionstyle = f"angle,angleA=0,angleB={ang}"
        annotation_text = '{}\n{}'.format(labels[i], autopct_values[i])
        
        # adjust_labels is to adjust the position of labels manually
        if adjust_labels is None:
            xytext = (1.35 * np.sign(x), 1.4 * y)
        else:
            xytext = (1.35 * np.sign(x), adjust_labels[i] * y)
            if adjust_labels[i] == 1.4:
                connectionstyle = f"angle,angleA=0,angleB={ang}"
            else:
                connectionstyle='arc3,rad=0'
            
        kw["arrowprops"].update({"connectionstyle": connectionstyle})
        ax.annotate(annotation_text, xy=(x, y), xytext=xytext,
                    horizontalalignment=horizontalalignment, **kw)
    return 
```

```python
def getXy(train, test, y_label, X_label=None):
    X_sub,X,y = [],[],[]
    if X_label is None:
        target = y_label
        features = [c for c in test.columns.tolist() if c!=target]
        X = train[features]
        X_sub = test[features]
    else:
        if len(X_label) == 1:
            features = X_label
            X = train[X_label].values.reshape(-1,1)
            X_sub = test[X_label].values.reshape(-1,1)
        else:
            features = X_label
            X = train[X_label].copy()
            X_sub = test[X_label].copy()
    y = train[y_label]
    return X_sub,X,y
```

```python
# feture engineering
# Kudos to https://www.kaggle.com/code/tetsutani/ps3e15-eda-ensemble-baseline
def create_features(df):
    """
    pressure: Pressure
    mass_flux: Mass flux
    D_e: Inner diameter
    D_h: Horizontal diameter
    length: Length
    chf_exp: Critical heat flux in experiments
    """
    # Mass flux and length product
    # df['mass_length_product'] = df['mass_flux'] * df['length']
    
    # Adiabatic surface area
    df['adiabatic_surface_area'] = df['D_e'] * df['length']
    
    # Surface area to horizontal diameter ratio
    df['surface_diameter_ratio'] = df['D_e'] / df['D_h']
    
    # Pressure to mass flux ratio
#     df['pressure_flux_ratio'] = df['pressure'] / df['mass_flux']
    
    # Combinations of features: create products or ratios between features
    # df['length_surface_ratio'] = df['length'] / df['adiabatic_surface_area']
    
    return df
```

```python
# Define columns
num_cols = train.select_dtypes(include=['float64']).columns.tolist()
num_cols.remove('x_e_out')
cat_cols = train.select_dtypes(include=['object']).columns.tolist()

# Numerical pipe
numerical_pipe = Pipeline(steps=[
    ('add_features', FunctionTransformer(create_features)),
    ('numerical', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

# categorical pipe
categorical_pipe = Pipeline(steps=[
    ('categorical', SimpleImputer(strategy='most_frequent')),
    ('OneHotEncod', OneHotEncoder(handle_unknown='ignore',sparse_output =False))
])
 
# Transform columns
preprocessor = ColumnTransformer(transformers=[
    ('numeric', numerical_pipe, num_cols),
    ('categorical', categorical_pipe, cat_cols)
], verbose_feature_names_out=False,).set_output(transform="pandas")

preprocessor.fit(X)
```

```python
import xgboost as xgb
# TRAIN
def runXgboost(X, y, X_sub, n_splits, params):

    # Define cross validation model
    kfold = KFold(shuffle = True, n_splits=n_splits, random_state=Seed)
    # Store feature importance
    features = X.columns
    importance_xgb = []
    # Store model scores and oof preds
    model_scores,rmse_scores, acc_scores = [],[],[]
    val_preds = np.zeros(len(X))
    test_preds = np.zeros(test.shape[0])

    for fold_idx, (train_idx, val_idx) in enumerate(kfold.split(X,y)):
        X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
        y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]
        # requires xgboost.DMatrix.
        dtrain = xgb.DMatrix(X_train, label=y_train, feature_names=features)
        dval = xgb.DMatrix(X_val, label=y_val, feature_names=features)
        watchlist = [(dtrain, 'train'), (dval, 'eval')]
        early_stop = xgb.callback.EarlyStopping(rounds=params['early_stopping_rounds'], metric_name='rmse', data_name='eval', save_best=True)
        model = xgb.train(params, dtrain, num_boost_round=params['n_estimators'], evals=watchlist, verbose_eval=False, callbacks=[early_stop])

        importance_xgb.append(model.get_score(importance_type='gain'))
        
        val_preds[val_idx] = model.predict(dval)
        test_preds += model.predict(xgb.DMatrix(X_sub, feature_names=features)) / (n_splits)
        
        # Evaluation different metrics
        metric = float(model.eval(dval).split(':')[1].strip())
        rmse = mean_squared_error(y_val, val_preds[val_idx], squared=False)
        model_scores.append(metric)
        rmse_scores.append(rmse)

        # Print metrics
        print(f'{Style.BRIGHT}{Fore.BLACK}Fold {Fore.RED}{fold_idx+1}{Fore.BLACK}: Model Metrics score = {Fore.RED}{metric:.5f}{Fore.BLACK}     RMSE scores = {Fore.RED}{rmse:.5f}{Style.RESET_ALL}')
    
    mean_scores = np.mean(model_scores)
    mean_rmse_scores = np.mean(rmse_scores)
    print(f'{"_" * 80}\n Mean Model score: {Style.BRIGHT}{Fore.RED}{mean_scores:.4f}{Fore.BLACK} (+/- {np.std(model_scores)**2})  Mean RMSE score: {Fore.RED}{mean_rmse_scores:.4f}{Style.RESET_ALL}')
    
    return mean_scores,mean_rmse_scores,test_preds, val_preds, importance_xgb
```
